```python
from pyspark.sql import SparkSession
import json
```

```python
sc = SparkSession.builder.appName("JSONReader").getOrCreate()
```

```python
rdd = sc.sparkContext.textFile('Amazon_Fashion.jsonl')
```

```python
reviews_rdd = rdd.map(lambda line: json.loads(line.strip()))
```

```python
print(reviews_rdd.take(5))
```

```python
rdd_2 = sc.sparkContext.textFile('meta_Amazon_Fashion.jsonl')
```

```python
meta_rdd = rdd_2.map(lambda line: json.loads(line.strip()))
```

```python
print(meta_rdd.take(5))
```

```python
reviews_df = reviews_rdd.toDF()
reviews_df.show()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, FloatType, ArrayType, MapType, IntegerType

# Define the schema based on the metadata fields
metadata_schema = StructType([
    StructField("main_category", StringType(), True),
    StructField("title", StringType(), True),
    StructField("average_rating", FloatType(), True),
    StructField("rating_number", IntegerType(), True),
    StructField("features", ArrayType(StringType()), True),
    StructField("description", ArrayType(StringType()), True),
    StructField("price", FloatType(), True),
    StructField("images", ArrayType(MapType(StringType(), StringType())), True),  # List of dictionaries
    StructField("videos", ArrayType(MapType(StringType(), StringType())), True),  # List of dictionaries
    StructField("store", StringType(), True),
    StructField("categories", ArrayType(StringType()), True),
    StructField("details", MapType(StringType(), StringType()), True),  # Dictionary
    StructField("parent_asin", StringType(), True),
    StructField("bought_together", ArrayType(StringType()), True)
])

# Convert `meta_rdd` to a DataFrame with the specified schema
metadata_df = sc.createDataFrame(meta_rdd, schema=metadata_schema)

# Display the DataFrame to verify
metadata_df.show(truncate=False)
```

```python
# Rename one of the `parent_asin` columns before joining to make them distinct
reviews_df = reviews_df.withColumnRenamed("parent_asin", "review_parent_asin")
metadata_df = metadata_df.withColumnRenamed("parent_asin", "meta_parent_asin")
```

```python
combined_df = reviews_df.join(metadata_df, reviews_df.review_parent_asin == metadata_df.meta_parent_asin, "inner")
```

```python
combined_df.show()
```

```python
# Average rating and total number of reviews
combined_df.select("average_rating", "rating").describe().show()

# Price distribution
combined_df.select("price").describe().show()
```

```python
# Count of each rating
rating_dist = combined_df.groupBy("rating").count().orderBy("rating")
rating_dist.show()
```

```python
from pyspark.sql.functions import year, avg

# Add a 'year' column from the 'timestamp' column
combined_df = combined_df.withColumn("review_year", year(combined_df["timestamp"].cast("timestamp")))

# Count of reviews per year
reviews_per_year = combined_df.groupBy("review_year").count().orderBy("review_year")
reviews_per_year.show()

# Average rating per year
avg_rating_per_year = combined_df.groupBy("review_year").agg(avg("rating").alias("avg_rating")).orderBy("review_year")
avg_rating_per_year.show()
```

```python
# Average rating by verified purchase
verified_rating = combined_df.groupBy("verified_purchase").agg(avg("rating").alias("avg_rating")).orderBy("verified_purchase")
verified_rating.show()
```

```python
# Count of each rating
rating_dist = combined_df.groupBy("rating").count().orderBy("rating")
rating_dist.show()
```

```python
# Top products by number of reviews, using explicit column references
most_reviewed_products = combined_df.groupBy("parent_asin", "title").count().orderBy("count", ascending=False)
most_reviewed_products.show(10, truncate=False)
```

```python
from pyspark.sql.functions import explode, split

# Explode features into individual words and count them
features_words = combined_df.withColumn("feature_word", explode(split("features", " ")))
features_count = features_words.groupBy("feature_word").count().orderBy("count", ascending=False)
features_count.show(20)
```

```python
# Top helpful reviews
top_helpful_reviews = combined_df.orderBy("helpful_vote", ascending=False).select("title", "text", "helpful_vote")
top_helpful_reviews.show(10, truncate=False)
```

```python
# Average rating and review count per store
store_performance = combined_df.groupBy("store").agg(
    avg("rating").alias("avg_rating"),
    count("rating").alias("review_count")
).orderBy("review_count", ascending=False)

store_performance.show(10, truncate=False)
```

```python
# Average rating per category
category_rating = combined_df.groupBy("main_category").agg(avg("rating").alias("avg_rating")).orderBy("avg_rating", ascending=False)
category_rating.show(10, truncate=False)
```
